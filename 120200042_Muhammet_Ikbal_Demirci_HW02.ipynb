```python
# this part define and print the shape, types, head, tail and the null variables of the data
def check_df(dataframe, head=5):
    print("##################### Shape #####################")
    print(dataframe.shape)
    print("##################### Types #####################")
    print(dataframe.dtypes)
    print("##################### Head #####################")
    print(dataframe.head(head))
    print("##################### Tail #####################")
    print(dataframe.tail(head))
    print("##################### NA #####################")
    print(dataframe.isnull().sum())

check_df(data1_data)
```

```python
# it defines as 1 that we want to use and defines 0 that we don't want to use
def one_hot_encoder(dataframe, categorical_cols, drop_first=False):
    dataframe = pandas.get_dummies(dataframe, columns=categorical_cols, drop_first=drop_first)
    return dataframe
```

```python
def grab_col_names(dataframe, cat_th=10, car_th=20):

  # dataframe takes dataframe and it means desired dataframe to retrieve variable names
  # cat_th takes integer and it means that class threshold value for variables that are numerical but categorical
  # car_th takes integer and it means that for the class threshold value for categorical but cardinal variables

    # cat_cols, cat_but_car

    cat_cols = [col for col in dataframe.columns if dataframe[col].dtypes == "O"]
    num_but_cat = [col for col in dataframe.columns if dataframe[col].nunique() < cat_th and
                   dataframe[col].dtypes != "O"]
    cat_but_car = [col for col in dataframe.columns if dataframe[col].nunique() > car_th and
                   dataframe[col].dtypes == "O"]
    cat_cols = cat_cols + num_but_cat
    cat_cols = [col for col in cat_cols if col not in cat_but_car]

  # cat_cols list categoric variable
  # num_cols list numerical variable list
  # cat_but_car categorical looking list of cardinal variables

    # num_cols
    num_cols = [col for col in dataframe.columns if dataframe[col].dtypes != "O"]
    num_cols = [col for col in num_cols if col not in num_but_cat]

    print(f"Observations: {dataframe.shape[0]}")
    print(f"Variables: {dataframe.shape[1]}")
    print(f'cat_cols: {len(cat_cols)}')
    print(f'num_cols: {len(num_cols)}')
    print(f'cat_but_car: {len(cat_but_car)}')
    print(f'num_but_cat: {len(num_but_cat)}')
    return cat_cols, num_cols, cat_but_car
```

```python
#prints how much are they to see us
cat_cols,num_cols,cat_but_car=grab_col_names(data1)
```

```python
# shows us the cat_cols and we can see that there are 3 of them as we mentioned upper block of code
cat_cols
```

```python
# encodes the categorical one, for example time-signature takes 3-4-5 variables only. 
#     If its 3 when we take a look computer says its a 1 but if isn't it says that 0 
#         and same for the 4-5 for every step and print it
data1=one_hot_encoder(data1,cat_cols,drop_first=True)

# after prints rename itself and helps for looking better
#tekrar yazdırdıktan sonra da kendisi adlandırarak daha net gözükmesini sağlıyo
data1
```

```python
# import the models from libraries that we need

from sklearn.ensemble import RandomForestClassifier,GradientBoostingClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from lightgbm import LGBMClassifier
from xgboost import XGBClassifier
from sklearn.svm import SVC
from sklearn.model_selection import train_test_split
```

```python
# we obtain x and y values to do train and test part,
#       x will be the other part of we select and y will be the we select
# we remove that we select the columns and print the others

x=data1.drop(["mode_1"],axis=1)
y=data1["mode_1"]
x.head()
```

```python
# creating an array for the models

from sklearn.model_selection import cross_validate
models=[("RF",RandomForestClassifier()),
      ( "GB",GradientBoostingClassifier()),
       ("DT",DecisionTreeClassifier()),
      ( "LG",LogisticRegression()),
      ( "KNC",KNeighborsClassifier()),
       ("LGBM",LGBMClassifier()),
       ("XGB",XGBClassifier())]

# return the array in "for" loop for see the results of every models
for name, model in models:

    cv_results = cross_validate(model, x, y, cv=10, scoring=["accuracy", "f1", "roc_auc", "precision", "recall"])
    print(f"########## {name} ##########")
    print(f"Accuracy: {round(cv_results['test_accuracy'].mean(), 4)}")
    print(f"Auc: {round(cv_results['test_roc_auc'].mean(), 4)}")
    print(f"Recall: {round(cv_results['test_recall'].mean(), 4)}")
    print(f"Precision: {round(cv_results['test_precision'].mean(), 4)}")
    print(f"F1: {round(cv_results['test_f1'].mean(), 4)}")
```

```python
# now we need to split the train and test parts from the data. It will be 20% for the test and 80% 
#         for the train because of we dont want to get an "overfitting" 
training_data, testing_data = train_test_split(data1_data, test_size=0.2, random_state=25)

# first we indicate the column numbers with "shape[0]" and get the output with f string method
print(f"No. of training examples: {training_data.shape[0]}")
print(f"No. of testing examples: {testing_data.shape[0]}")

# and we call it
testing_data
```

```python
# we split the train and test datas again. 

# we remove the "move" column and assign to the x_train
x_train=training_data.drop('mode',axis=1)
#"move" column and assign to the y_train
y_train=training_data['mode']

# we remove the "move" column and assign to the x_test
x_test = testing_data.drop('mode', axis=1)
# "move" column and assign to the y_test
y_test = testing_data['mode']


#Shapes of x_train,y_train and test data
x_train.shape, y_train.shape, testing_data.shape
```

```python
from sklearn.ensemble import RandomForestClassifier

# we cretea an object from "RandomClassifier" class
rfc = RandomForestClassifier(n_estimators=100, max_depth=10, random_state=42)

# train the model with fit method
rfc.fit(x_train, y_train)

# take a guess from test datas and take a guess on x_test datas and compare with y_test datas
y_pred = rfc.predict(x_test)

from sklearn.metrics import accuracy_score
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy is:", accuracy)
```

**Hyperparameter Optimization**

```python
# trying to find best Hyperparameter combination

from sklearn.model_selection import GridSearchCV
rf_model=RandomForestClassifier(random_state=17)
rf_params={"max_depth":[5,8,None],
           "max_features":[3,5,7,"auto"],
           "min_samples_split":[2,5,8,15,20],
           "n_estimators":[100,200,500]}

# GridSearchCV trains the model with givin parameters and evaluate the results
#     when finds the best hyperparameters found again creates an model and train with best hyperparameters
#         we measure the performance of the model with "cross_validate" method "f1" and "roc_auc" values are calculated and printed as an average.
rf_best_grd=GridSearchCV(rf_model,rf_params,cv=5,n_jobs=-1,verbose=True).fit(x,y)
print(rf_best_grd.best_params_)
rf_final=rf_model.set_params(**rf_best_grd.best_params_,random_state=17).fit(x,y)
cv_results=cross_validate(rf_final,x,y,cv=5,scoring=["accuracy","f1","roc_auc"])

print(cv_results["test_accuracy"].mean())
print(cv_results["test_f1"].mean())
print(cv_results["test_roc_auc"].mean())

#### RF before ####   #### RF now ####
# Accuracy: 0.6047      Accuracy: 0.6513
# Auc: 0.6611           Auc: 0.6450
# Recall: 0.6091
# Precision: 0.6455
# F1: 0.6156            F1: 0.6782
```
