# Visualizing Dataset

```
# Generating synthetic dataset

def func(X): return X**2

X = np.arange(0, 100)
y = func(X) + np.random.randint(10, size=(100)) * np.random.normal(100)
data = zip(X, y)
```

```
syn_data = pandas.DataFrame(data, columns=['X', 'y'])
print(syn_data.head())
```

## Boxplot

```
setosa = iris_data[iris_data['species'] == 0]
versicolor = iris_data[iris_data['species'] == 1]
virginica = iris_data[iris_data['species'] == 2]
```

```
feature = 'sepal_length'

bp_data = [setosa[feature], versicolor[feature], virginica[feature]]
plt.boxplot(bp_data, labels=['setosa', 'versicolor', 'virginica'])
plt.show()
```

```
sns.boxplot(x='species', y='sepal_length', data=iris_data)
plt.show()
```

### _What is the difference between box and bar plot?_

```
sns.barplot(x='species', y='sepal_length', data=iris_data)
plt.show()
```
